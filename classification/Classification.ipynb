**Train a model:**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

**Machine Learning Models Used**

Common models used with this dataset:
- KNeighborsClassifier – Classifies data based on the nearest neighbors.
- SVC (Support Vector Classifier) – Separates classes using the best hyperplane.
- MultinomialNB – A Naive Bayes model mainly used for discrete features like counts.
- DecisionTreeClassifier – Predicts outcomes using a tree-like decision structure.
- AdaBoostClassifier – Combines weak models sequentially to improve performance.
- GradientBoostingClassifier – Builds models step-by-step to reduce prediction errors.
- XGBClassifier – An optimized gradient boosting algorithm with high performance.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
x=XGBClassifier()
gr=GradientBoostingClassifier()
ad=AdaBoostClassifier()
d=DecisionTreeClassifier(random_state=1)
g=MultinomialNB()
svc=SVC()
knn=KNeighborsClassifier()
l=[gr,ad,d,g,svc,knn]
for i in l:
    print(i)
    i.fit(X_train,Y_train) 
    y_pred=i.predict(X_test)
    print(classification_report(Y_test,y_pred))
```

- AdaBoostClassifier – 97% accuracy (Best Model)
- GradientBoostingClassifier – 96% accuracy
- DecisionTreeClassifier – 95% accuracy
- KNeighborsClassifier – 94% accuracy
- SVC – 90% accuracy
- MultinomialNB – 87% accuracy

**Conclusion: AdaBoostClassifier performed the best for this dataset.**

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier()
ad.fit(X_train, Y_train)
y_pred_ad = ad.predict(X_test)
print("AdaBoost Classifier:")
print(classification_report(Y_test, y_pred_ad))
```
